```python
import torch
import torch.nn as nn
import numpy as np
from data_load import data_load
from util import normalize_func, train_test_split, train_test_split_seq
import random
from train import train_lstm, train_seq, train_seq_attn, train_seq_attn2, train_transformer
from predict import predict, predict_seq, predict_trans
```

```python
#LSTM /GRUのやり方
data_type = 'ALL'
#ALL, 10Area, 9Area, states
ili_weighted, date = data_load(data_type, 1)

test_num = 52
validate_num = 52

ili_weighted, minmax = normalize_func(np.array(ili_weighted).reshape(-1,1), validate_num, test_num)

time_window = 52
future_step = 2
train_d, val_d, test_d =train_test_split_seq(ili_weighted, validate_num, test_num, time_window, future_step = future_step)
train_date, val_date, test_date =train_test_split_seq(date, validate_num, test_num, time_window, future_step = future_step)
train_input = torch.Tensor(train_d[0])
train_output = torch.Tensor(train_d[1])
valid_input = torch.Tensor(val_d[0])
valid_output = torch.Tensor(val_d[1])
test_input = torch.Tensor(test_d[0])
test_output = torch.Tensor(test_d[1])
input_size = train_input.shape[2]

input_size = train_input.shape[2]
hidden_size = 30
seq_len = train_input.shape[1]
lr = 0.005
epochs = 20
enc_layers = 3
dec_layers = 3
n_heads = 3
en_pf_size = 10
de_pf_size = 10
batch_n = 10
model = train_transformer(train_input, train_output, valid_input, valid_output, hidden_size, enc_layers, dec_layers, n_heads, en_pf_size, de_pf_size, epochs, batch_n = batch_n)
```

```python
def predict_trans(model, test_input, test_output):
    predict_ = []
    model.eval()
    with torch.no_grad():
        for test_i, test_o in zip(test_input, test_output):
            test_i = test_i.unsqueeze(0)
            test_o = test_o.unsqueeze(0)
            predict_.append(model(test_i, test_o)[0].squeeze(0).squeeze(1).numpy())
    return np.array(predict_)
```

```python
pred = np.array(predict_trans(model, test_input, test_output))
```

```python
from matplotlib import pyplot as plt
plt.plot(pred[:,-1])
plt.plot(test_output[:,-1])
```
